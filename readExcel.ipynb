```python
import pandas as pd

# We need the following packages installed:
# openpyxl - xlsx files
# xlrd - xls files
# More information here: https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html
```

```python
# Minimal example
xls = pd.ExcelFile('data.xlsx'); # Read Excel file into object.
df = pd.read_excel(xls); # put object into data frame. This only takes the first sheet.
df # show
```

```python
# Read sheets

xls = pd.ExcelFile('data.xlsx'); 
df = pd.read_excel(xls, sheet_name='Sheet2'); # put second sheet into data frame
df # show
```

```python
# Sheet3 does not have column titles, so add our own

xls = pd.ExcelFile('data.xlsx'); 
df = pd.read_excel(xls, sheet_name='Sheet3', header=None, names=['Column E', 'Column F']); # put second sheet into data frame
df # show
```

```python
# Use first column as index labels in Sheet4

xls = pd.ExcelFile('data.xlsx'); 
df = pd.read_excel(xls, sheet_name='Sheet4', index_col=0); # put second sheet into data frame
df # show
```

```python
# Write to excel file
df.to_excel('out.xlsx')
```
